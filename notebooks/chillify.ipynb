```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
```

```python
data1 = pandas.read_csv('../losses/chillify_translator/losses.csv')
c_right = data1['C_A_right'] + data1['C_B_right']
c_wrong = data1['C_A_wrong'] + data1['C_B_wrong']
d = data1['D_A'] + data1['D_B']
c_right[::].plot(ylim=(-1, 2))
c_wrong[::].plot(ylim=(-1, 2))
d[::].plot(ylim=(0, 15))
plt.title('Chillify dataset - UMTN')
plt.savefig('chillify_umtn.png')
```

```python
l = len(d)
mid = l // 2 + 25
print(d[50:mid].mean() - d[mid:l].mean())
```
