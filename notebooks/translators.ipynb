```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
```

```python
data1 = pandas.read_csv('../losses/p2g_translator_3/losses.csv')
c_right1 = data1['C_A_right'] + data1['C_B_right']
c_wrong1 = data1['C_A_wrong'] + data1['C_B_wrong']
# c1 = data1['C_A'] + data1['C_B']

d1 = data1['D_A'] + data1['D_B']
c_right1[::].plot(label='Classification Loss')
c_wrong1[::].plot(label='Domain Confusion Loss')
# c1.plot()
d1[::].plot(ylim=(0, 12), label='Reconstruction Loss')
plt.title('piano2guitar dataset - UMTN')
plt.legend()
plt.savefig('p2g_umtn.png')
```

```python
data1['D_A'].plot(label='Piano Decoder')
data1['D_B'].plot(label="Guitar Decoder")
plt.legend()
plt.title('Decoder reconstruction loss')
plt.savefig('decoder.png')
plt.show()
```

```python
l = len(d1)
mid = l // 2 + 15
print(d1[30:mid].mean() - d1[mid:l].mean())
```
