```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas
```

```python
results = pandas.read_csv('../results.csv')
```

```python
results.groupby(['section', 'qtype']).mean()
```

```python
results.groupby(['question_id', 'qtype']).mean()
```

```python
results.groupby(['section','is_musician', 'qtype']).mean()
```

```python
for qtype in ['content', 'style']:
    for i in range(16):
        results[(results.question_id == i+1) & (results.qtype == qtype)]['answer'].value_counts().sort_index().plot.bar(width=1, rot=1, figsize=(3,3))
        plt.title('%s Q. %s - %s' % ('piano2guitar' if (i // 8) == 0 else 'guitar2piano', i% 8 + 1, 'Style' if qtype == 'style' else 'Content'))
        plt.xticks(range(5))
        plt.xlim(-1,5)
        plt.savefig('figures/res_%s_%s_%s.png' % ((i // 8) + 1, (i % 8) + 1, qtype))
        plt.show()
```

```python
for qtype in ['content', 'style']:
    for section in range(1, 3):
        results[(results.section == section) & (results.qtype == qtype)]['answer'].value_counts().sort_index().plot.bar(width=1, rot=1, figsize=(5,5))
        plt.title('%s - %s' % ('piano2guitar' if section == 1 else 'guitar2piano', 'Style' if qtype == 'style' else 'Content'))
        plt.savefig('sec_%s_%s.png' % (section, qtype))
        plt.show()
```
